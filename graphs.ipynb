```python
from matplotlib import pyplot as plt
```

```python
#graph for no time

threshold = [1,2,3,4,5]
accuracy = [100,100,90,70,55]

plt.plot(threshold, accuracy)
plt.xlabel('Threshold')
plt.ylabel('Accuracy')
plt.title('No-time')
plt.show()
```

```python
#graph for constant time
threshold_constant = [0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50, 0.55, 0.60, 0.65, 0.70, 0.75, 0.80, 0.85, 0.90, 0.95, 1.00]
accuracy_constant = [50, 50, 50, 50, 50, 55, 60, 65, 75, 85, 85, 90, 95, 95, 100, 50, 50, 50, 50, 50]

plt.plot(threshold_constant, accuracy_constant)
plt.xlabel('Threshold')
plt.ylabel('Accuracy')
plt.title('Constant-time')
plt.show()
```

```python
#graph for random time

threshold = [0.05,0.10,0.15,0.20,0.25]
accuracy = [55,45,50,65,60]

plt.plot(threshold, accuracy)
plt.xlabel('Threshold')
plt.ylabel('Accuracy')
plt.title('Random-time')
plt.show()
```
